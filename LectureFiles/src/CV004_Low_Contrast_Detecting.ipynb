# **Low Contrast Detecting(저대비 이미지 감지)**

- 대비(Contrast)가 낮은 이미지
    - 밝은 부분과 어두운 부분의 구분이 최소화되어 있음
    - 물체의 경계가 시작되는 위치와 장면의 배경이 시작되는 위치를 구분하기 어려움

- 저대비 이미지 감지 사용
    - 이미지 처리 파이프라인에 충분하지 않은 이미지를 프로그래밍 방식으로 감지할 수 있음

- 처리 순서
    - 필요한 모든 라이브러리를 가져옴
    - 회색 음영 모드에서 이미지 읽기
    - 제공된 임계값을 사용하여 이미지가 저대비 이미지인지 고대비 이미지인지 확인
        - skimage.exposure의 is_low_contrast 함수 사용

- 필요한 라이브러리 가져오기

```python
from skimage.exposure import is_low_contrast

img = cv2.imread("./low_contrast.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

if(is_low_contrast(gray,0.35)):
    cv2.putText(img, "low contrast image", (5, 25),
                cv2.FONT_HERSHEY_SIMPLEX, 0.8,
    (0,0,0), 2)
else:
    cv2.putText(img, "high contrast image", (5, 25),
                cv2.FONT_HERSHEY_SIMPLEX, 0.8,
    (0,0,0), 2)

cv2_imshow(img)
cv2.imwrite("output.jpg",img)
```

```python
img = cv2.imread("./high_contrast.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
if(is_low_contrast(gray,0.35)):
    cv2.putText(img, "low contrast image", (5, 25),
                cv2.FONT_HERSHEY_SIMPLEX, 0.8,
    (0,0,0), 2)
else:
    cv2.putText(img, "high contrast image", (5, 25),
                cv2.FONT_HERSHEY_SIMPLEX, 0.8,
    (0,0,0), 2)

cv2_imshow(img)
cv2.imwrite("output.jpg",img)
```
