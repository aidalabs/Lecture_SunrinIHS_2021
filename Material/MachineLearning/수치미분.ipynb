## **1-1. 1차 방정식과 그래프**

- 변수와 상수(1차 방정식 기준)를 이용해서 직선을 그려보자
  - $y = 1.5x$

- 참고
  - np.linspace() : 숫자로 된 시퀀스를 생성하는 함수. NumPy Array로 구성된 균등한 간격을 둔 시퀀스를 생성함
  - 시작점과 종점을 정하고, 해당 간격 내의 브레이크 포인트의 총 갯수를 정하면 np.linspace 함수는 해당 간격 내에 일정한 간격을 둔 브레이크 포인트 갯수만큼의 시퀀스를 반환함. 기본 갯수는 50개
```
np.linspace(시작, 끝, 갯수)
```

````python
import numpy as np
import matplotlib.pyplot as plt

a = 1.5  # a: 상수
x = np.linspace(-1,1)  # x: 변수 -1부터 1의 범위
y = a * x  # y: 변수

plt.plot(x, y)
plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()
plt.show()
````

## **1-3. 함수와 그래프**

- 함수를 이용하여 제곱근의 그래프를 그려보자
  - $y = f(x) = x^3$

````python
import numpy as np
import matplotlib.pyplot as plt

def my_func(x):
    return np.sqrt(x) + 1

x = np.linspace(0, 4)
y = my_func(x)  # y = f(x)

plt.plot(x, y)
plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()
plt.show()
````

- 함수를 이용하여 다항식의 그래프를 그려보자
  - $y = x^3 - 2x^2 - 3x + 4$

````python
import numpy as np
import matplotlib.pyplot as plt

def my_func(x):  
    return x**3 - 2*x**2 - 3*x + 4

x = np.linspace(-2, 2)
y = my_func(x)  # y = f(x)

plt.plot(x, y)
plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()
plt.show()
````

````python
# 연립방정식 풀기
# sympy 라이브러리를 불러오고, 사용할 기호 변수 x, y를 선언한다.
from sympy import Symbol, solve
x=Symbol('x')
y=Symbol('y')

# 방정식을 풀기 위해서는 “(방정식) = 0”으로 만들기 위해 모든 식을 좌변으로 이항시켜준 후, equation1과 equation2로 변수화해준다
equation1 = 3 * x + y - 2
equation2 = x - 2 * y - 3

# 방정식을 풀기 위해 sympy에 내장된 solve() 함수를 사용한다. solve() 함수를 사용하기 위해 equation을 차례로 입력해주면 방정식을 풀어서 결과를 반환한다.
solve((equation1, equation2), dict=True)  # dict 옵션은 해를 사전 형태로 반환한다.
````

# **2. 선형대수**

- 벡터 공간, 벡터, 선형 변환, 행렬, 연립 선형 방정식 등을 연구하는 대수학의 한 분야
  - 벡터: 벡터 공간의 원소. 방향과 크기를 가지고 있는 힘의 크기를 나타냄
  - 벡터 연산: 두 벡터끼리의 합, 혹은 벡터와 스칼라(크기만 있고 방향성은 없는 성분)사이의 곱이 벡터의 기본 연산
  - 벡터 공간: 벡터의 기본 연산을 만족하는 모든 벡터의 모음
  - 차원: 흔히 평면을 2차원, 공간을 3차원이라고 부른다. 이때 차원을 구성하는 각각의 요소(3차원의 경우 x,y,z)는 서로 독립적인데 이에 대한 개념을 확장한 것이 바로 선형대수학의 차원임
  - 행렬: 여러 개의 숫자들을 직사각형의 모양으로 한데 묶어 나타낸 성분. 벡터를 하나의 행 혹은 하나의 열로 구성된 행렬로 볼 수도 있지만 이것이 수학적으로 행렬의 엄밀한 정의는 아님
####<br>
- 현대의 선형 대수학은 특히 벡터 공간이 주 연구 대상이며 자연과학과 공학에도 널리 활용됨
####<br>
- 어떤 함수가 선형(Linear)일 때 그 함수의 성질을 배우는 것이며, 선형은 다음과 같은 두 개의 식으로 정의할 수 있음
  - $f(kx)=kf(x)$
  - $f(x_1 + x_2) = f(x_1 )+f(x_2 )$
####<br>
- 대표적인 선형 함수
  - $y=Wx$ 와 같은 정비례함수, 미분, 적분, 회전 변환, 확대, 축소 변환
####<br>
- 대표적인 비선형 함수: 이차함수, 삼각함수, 로그함수, 지수함수
####<br>
- 선형대수 이론에 따르면
  - 임의의 선형함수는 행렬로 표현할 수 있으며, 반대로 행렬은 어떤 선형함수에 대응함
  - 행렬에 대한 성질만 익히면 모든 선형 함수의 공통 성질을 알 수 있음
  - 필요에 따라 비선형함수도 근사를 통해 선형대수를 적용할 수 있음

## **2-2. 벡터의 내적과 놈(Norm)**

- 벡터:
- 놈: 벡터의 크기를 나타내는 값 또는 함수. AI에서는 L2놈과 L1놈이자주 사용됨
  - L1 Norm : 벡터 p, q가 있을 때, p, q의 각 원소들의 차이의 절대값의 합
  - L2 Norm : 벡터 p, q의 유클리디안 거리(직선 거리)
  - L1 Norm과 L2 Norm의 차이
  image.png

  L1 Norm은 빨간선, 파란선, 노란선 모두 가능하지만 L2 Norm은 초록색 선만 가능함

````python
# 벡터의 내적 계산
import numpy as np

a = np.array([1, 2, 3])
b = np.array([3, 2, 1])

print("--- dot()함수 ---")
print(np.dot(a, b))  # dot()함수에 의한 내적 
print("--- 곱의 총합 ---")
print(np.sum(a * b))  # 곱의 총합에 의한 내적
````

````python
# 벡터의 놈 계산
import numpy as np

a = np.array([1, 1, -1, -1])

print("--- L2놈 ---")
print(np.linalg.norm(a))  # L2놈(디폴트)
print("--- L1놈 ---")
print(np.linalg.norm(a, 1))  # L1놈
````

````python
# 주어진 벡터 a와 벡터 b의 내적과 벡터 a의 L2놈과 L2놈 표시
import numpy as np

a = np.array([1, -2, 2])
b = np.array([2, -2, 1])

print("--- 내적 ---")
print(np.dot(a, b))

print("--- L2놈 ---")
print(np.linalg.norm(a))

print("--- L1놈 ---")
print(np.linalg.norm(a, 1))
````

## **2-5. 행렬식과 역행렬**

- 행렬식이 0이면 역행렬이 존재하지 않음
- 역행렬은 인공지능에서 변수 사이의 상관관계를 알아보는 회귀분석에 사용됨

````python
# 단위행렬 만들기
import numpy as np

print(np.eye(2))  # 2×2의 단위 행렬
print()
print(np.eye(3))  # 3×3의 단위 행렬
print()
print(np.eye(4))  # 4×4의 단위 행렬
````

````python
# 행렬식 구하기
import numpy as np

a = np.array([[1, 2],
              [3, 4]])
print(np.linalg.det(a))  # 행렬식이 0이 되지 않는 경우 --> 역행렬 존재

b = np.array([[1, 2],
              [0, 0]])
print(np.linalg.det(b))  # 행렬식이 0이 되는 경우 --> 역행렬 존재하지 않음
````

````python
# 역행렬 구하기
import numpy as np

a = np.array([[1, 2],
              [3, 4]])
print(np.linalg.inv(a))  # 역행렬

b = np.array([[1, 2],
              [0, 0]])
# print(np.linalg.inv(b))  # 역행렬이 존재하지 않으므로 에러가 난다
````

## **2-6. 선형 변환**

````python
# 벡터를 화살표로 그려보자
import numpy as np
import matplotlib.pyplot as plt

# 화살표를 그리는 함수
def arrow(start, size, color):
    plt.quiver(start[0], start[1], size[0], size[1], angles="xy", scale_units="xy", scale=1, color=color)

# 화살표의 시작점
s = np.array([0, 0])  # 원점

# 벡터
a = np.array([2, 3])  # 세로 벡터를 나타낸다

arrow(s, a, color="black")

# 그래프 표시
plt.xlim([-3,3])  # x의 표시 범위
plt.ylim([-3,3])  # y의 표시 범위
plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()
plt.gca().set_aspect("equal")  # 가로세로비를 같게
plt.show()
````

# **3. 극한과 미분**

- 극한: 함수에서의 변수값을 어떤 값에 가깝게 할때, 함수의 값이 한없이 가까워지는 값
- $ \lim\limits_{x \to 0}y=\lim\limits_{x \to 0}(x^2 +1)=1 $ : x를 한없이 0에 가깝게 하면 y가 한없이 1에 가까워진다.
  - $x=2$ 일 때 $y=5$
  - $x=1$ 일 때$ y=2$
  - $x=0.5$ 일 때 $y=1.25$
  - $x=0.1$ 일 때 $y=1.01$
  - $x=0.01$ 일 때 $y=1.0001$
####<br>
- 미분: 함수$f(x)$로부터 도함수 $f'(x)$를 얻는 것을 함수 $f(x)$를 미분한다고 함
  - 함수 $y=f(x)$에서의 $x$의 미소한 변화량을 $x$라고 하면 $x$를 $\Delta x$만큼 변화시킬 때의 y의 값은 $y=f(x+ \Delta x)$이며, 이 때, $y$의 미소한 변화량은 $\Delta y=f(x+ \Delta x)-f(x)$
####<br>
  - 따라서 $y$의 미소한 변화 $\Delta y$와 $x$의 미소한 변화 $\Delta x$의 비율은 ${\Delta y \over \Delta x}={f(x+ \Delta x)-f(x) \over \Delta x}$
####<br>
  - 여기에서 $\Delta x$의 값이 0에 한없이 가까워지는 극한을 적용하면
  - $f'(x)= \lim\limits_{\delta x \to 0} {f(x+ \Delta x)-f(x) \over \Delta x}={df(x) \over dx}={d \over dx} f(x)$
####<br>
  - 함수의 변수가 하나뿐인 경우, 이와 같은 1변수 함수에 대한 미분을 상미분,
  - 함수의 변수가 두 개 이상인 경우, 두 개의 변수 중 하나의 변수에 대해서만 미분하는 것을 편미분이라고 함

## **3-3. 수치미분**

- 컴퓨터는 극한을 취급할 수 없음 $\rightarrow$ 극한과 비슷한 값으로 대체하여 계산
- 이런 미세한 차이를 이용하여 함수의 변화량을 구하는 방법을 수치미분 이라고 함
  - 수치 미분은 매우 작은 값을 사용하여 진짜 미분값을 근사함
  - 따라서 어쩔 수 없이 결과 값에 오차가 포함됨
  - 이러한 근사 오차를 줄이는 방법 $\rightarrow$ 중앙차분(centered difference)
  - 중앙 차분: $f(x)$와 $f(x+h)$의 차이를 구하는 대신 $f(x-h)$와 $f(x+h)$의 차이를 구함

````python
import numpy as np
import matplotlib.pylab as plt
 
def numerical_differentiation(f, x):    # 수치미분
    delta_x = 1e-7    # 0.0000001
    # return (f(x+delta_x) - f(x)) / delta_x            # 전진 차분
    return (f(x+delta_x) - f(x-delta_x)) / (delta_x*2)  # 중앙 차분
    # return (f(x) - f(x-delta_x)) / delta_x            # 후진 차분
 
 
def f1(x):
    return x**2
 
def f2(x):
    return x**3 + x**2
 
x = np.arange(0,10,0.01)
 
plt.plot(x, f1(x), label="f1(x)")
plt.plot(x, numerical_differentiation(f1,x), label="f1(x)_derivative")
 
plt.plot(x, f2(x), label="f2(x)")
plt.plot(x, numerical_differentiation(f2,x), label="f2(x)_derivative")
 
plt.ylim(0, 300)
plt.legend()
plt.show()
````

````python
# 수치미분과 해석미분의 차이 비교
import numpy as np
import matplotlib.pylab as plt
 
def numerical_differentiation(f, x):    # 수치미분
    delta_x = 1e-7    # 0.0000001
    # return (f(x+delta_x) - f(x)) / delta_x            # 전진 차분
    return (f(x+delta_x) - f(x-delta_x)) / (delta_x*2)  # 중앙 차분
    # return (f(x) - f(x-delta_x)) / delta_x            # 후진 차분
 
 
def f(x):
    return x**3 + x**2
 
def f_analytic(x):
    return 3*(x**2) + 2*x
 
x = np.arange(0,10,0.01)

plt.figure(figsize=(15, 10))
plt.plot(x, f(x), label="f(x)")
plt.plot(x, numerical_differentiation(f,x), lw=4, color='g', label="f(x)_numerical_differentiation")
plt.plot(x, f_analytic(x), linestyle='--', lw=4, color='r', label="f(x)_analytic_differentiation")
 
plt.ylim(0, 300)
plt.legend()
plt.show()
````

- 합성함수 미분하기
  - $y=(e^{x^2})^2$
  - 전체 수식을 f라는 함수를 만들고 그 내부에서 함수들을 묶어서 사용
  - 이런 방식으로 복잡하게 조합된 함수도 자동 미분계산 가능함(미분 가능한 함수의 경우)

````python
# 합성 함수 미분하기

import numpy as np

class Variable:
    def __init__(self, data):
        self.data = data

class Function:
    def __call__(self, input):
        x = input.data
        y = self.forward(x)
        output = Variable(y)
        return output

    def forward(self, x):
        raise NotImplementedError()

class Square(Function):
    def forward(self, x):
        return x**2

class Exp(Function):
    def forward(self, x):
        return np.exp(x)

def numeriacla_diff(f, x, eps=1e-4):
    x0 = Variable(x.data - eps)
    x1 = Variable(x.data + eps)
    y0 = f(x0)
    y1 = f(x1)
    return (y1.data - y0.data) / (2 * eps)

def f(x):
    A = Square()
    B = Exp()
    C = Square()
    return C(B(A(x)))

x = Variable(np.array(0.5))
dy = numeriacla_diff(f, x)
print(dy) 
````

- 수치미분의 단점
  - 결과에 오차가 포함되어 있다
    - 대부분의 경우 오차는 매우 작지만 어떤 계산이냐에 따라 커질 수도 있음
  - 계산량이 많다
    - 변수가 여러 개인 계산을 미분할 경우, 변수 각각을 모두 미분해야 함
    - 특히 신경망, 딥러닝 모델에서는 매개 변수가 수백만 개 이상을 사용하는 경우가 매우 흔하므로 모든 것을 수치미분으로 구하는 것은 비현실적이다
    - 이런 이유로 등장한 것이 역전파 알고리즘

## **3-3. 네이피어 수(e)와 자연대수**

  - ${\partial y \over \partial x}=\lim\limits_{\Delta x \to 0}{e^{x+\Delta x} -e^x \over \Delta x}=e^x$ 와 같이 미분해도 식이 변하지 않는 특징을 가짐. 이 성질때문에 네이피어 수는 수학적으로 다루기 쉽고, 인공지능에서의 여러가지 수식에서 사용됨
  - $y=exp(x)$ 와 같이 표기할 수도 있음

- 자연대수

  - 특히 $a$가 네이피어수 $e$인 경우, $log_a x$를 자연대수라고 부름
  - $y=log_e x$ ( --> $e$를 $y$제곱하면 $x$가 됨)
  - 자연대수: $e$를 몇 제곱하면 $x$가 되는가를 나타냄

## 3-4. 시그모이드 함수

- 신경망, 딥러닝 모델에서 사용되는 시그모이드 함수(네이피어 수를 사용함)
- $y={1 \over {1+exp(-x)}}$
- 시그모이드 함수는 x의 값이 작아지면 0에 가까워지고, x의 값이 커지면 1에 가까워짐, 또한 그 도함수는 x가 0일때 최댓값인 0.25를 가지고, 0으로부터 멀어지면 0에 가까워짐
- 시그모이드 함수를 이용하여 입력을 0과 1 사이의 연속적인 출력으로 변환할 수 있음
- 신경망, 딥러닝에서는 인공적인 신경세포의 흥분 정도를 나타내는 활성화함수로 사용됨
- 시그모이드 함수의 도함수는 학습을 통하여 신경망을 최적화시키는 백프로퍼게이션 알고리즘에 사용됨

````python
# 시그모이드 함수와 그 도함수의 그래프 그리기
import numpy as np
import matplotlib.pylab as plt

def sigmoid_function(x):  # 시그모이드 함수
    return 1/(1+np.exp(-x))

def grad_sigmoid(x):  # 시그모이드 함수의 도함수
    y = sigmoid_function(x)
    return (1-y)*y

x = np.linspace(-5, 5)
y = sigmoid_function(x)
y_grad = grad_sigmoid(x)

plt.plot(x, y, label="y")
plt.plot(x, y_grad, label="y_grad")
plt.legend()

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()

plt.show()
````

## **3-5. 최급강하법(경사하강법)**

- 함수의 미분값(기울기)을 바탕으로 최솟값 등의 탐색을 시행하는 알고리즘
- 경사법의 일종으로, 가장 급한 방향으로 강하하도록 해서 최솟값을 탐삭함
<br /><br />
- 알고리즘
  - 다변수 함수 $f(\vec x)$의 최솟값 탐색 : $f(\vec x)=f(x_1 , x_2 , \cdots , x_i , \cdots , x_n )$
  - $\vec x$의 초깃값을 적당히 정한 후에 다음 식에 의거한 모든 요소를 갱신 : $x_i \leftarrow x_i - \eta{\partial f(\vec x) \over \partial x_i}$
  - $\eta$는 학습계수로 불리는 상수, $x_i$의 갱신 속도 결정
  - 이 식에 의해 기울기가 클수록(경사가 급할수록) $x_i$값은 크게 변경됨
  - 이것을 $f(\vec x)$가 변화하지 않게 될 때까지(기울기가 0이 될 때까지) 반복함으로써 $f(\vec x)$의 최솟값을 구함

````python
# 최급강하법에 의해 함수의 최솟값을 구해보자
import numpy as np
import matplotlib.pyplot as plt

def my_func(x):  # 최솟값을 구하는 함수
    return x**2 - 2*x

def grad_func(x):  # 도함수
    return 2*x - 2

eta = 0.1  # 학습계수
x = 4.0  # x에 초깃값을 설정
record_x = []  # x의 기록
record_y = []  # y의 기록
for i in range(20):  # 20회 x를 갱신한다
    y = my_func(x)
    record_x.append(x)
    record_y.append(y)
    x -= eta * grad_func(x)  # （식1）

x_f = np.linspace(-2, 4)  # 표시 범위
y_f = my_func(x_f)  

plt.plot(x_f, y_f, linestyle="dashed")  # 함수를 점선으로 표시
plt.scatter(record_x, record_y)  # x와 y의 기록을 표시

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()

plt.show()
````
